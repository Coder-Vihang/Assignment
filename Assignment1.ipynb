```python
import pandas as pd
```

```python
import fitz

file = fitz.open("/content/keppel-corporation-limited-annual-report-2018.pdf")
```

```python
for page_number, page in enumerate(file.pages(), start=1):
  text= page.getText()
  txt=open(f"pdf_page_wise{page_number}.txt","a")
  txt.writelines(text)
  
  txt.close()

  
```

```python
import xlwt
from itertools import chain

wb = xlwt.Workbook()
sheet = wb.add_sheet("New Sheet1")
```

```python
for page_number, page in enumerate(file.pages(), start=1):
  data=[]
  with open(f"pdf_page_wise{page_number}.txt") as f:
    for line in f:
      data.append([word for word in line.split(" ") if word])
      list1=[]
      list1 = list(chain.from_iterable(data))
    for i in range(len(list1)):
      sheet.write(i, page_number, list1[i])
wb.save('assignment1.xls')

  
```
